```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def load(method: str, vel: float):
    path_real = f'data/hw3/{method}_{vel}_real.npy'
    path_history = f'data/hw3/{method}_{vel}_history.npy'

    real = np.load(path_real, allow_pickle=True)
    history = np.load(path_history, allow_pickle=True)

    return real, history
```

```python
trajectory = []
for t in np.linspace(0, 40, 200):
    trajectory.append([3 * np.sin(0.1 * t), t])

trajectory = np.array(trajectory)
```

```python
def plot(real, history, title):
    plt.figure()
    plt.title(title)
    plt.plot(real[:, 0], real[:, 1], label='real position of car')
    plt.plot(history[:, 0], history[:, 1], label='calculated position of car')
    plt.plot(trajectory[:, 0], trajectory[:, 1], label='trajectory')

    plt.legend()
    plt.grid()
```

```python

plot(*load('pure_pursuit', 0.5), 'Pure pursuit 0.5 m/s')
```

```python

plot(*load('pure_pursuit', 5), 'Pure pursuit 5 m/s')
```

```python

plot(*load('stanley_control', 0.5), 'Stanley control 0.5 m/s')
```

```python

plot(*load('stanley_control', 5), 'Stanley control 5 m/s')
```
